```python
from cryptography.fernet import Fernet
import os
from dotenv import load_dotenv
from multiprocesspandas import applyparallel

load_dotenv()
```

```python
import pandas as pd

# Create a dataframe with 500000 entries
numbers = []
for i in range(500000):
    numbers.append({"number": f"923244{i}413{i+1}"})

df = pd.DataFrame(numbers)
df.head()
```

```python
def encrypt_entries(phone_number):
    return fernet.encrypt(phone_number.encode()).decode()
```

```python
fernet = Fernet(os.getenv("FERNET_KEY"))

from time import time

start = time()
for i in range(500000):
    encrypted_phone = fernet.encrypt("923244415089".encode()).decode()
f"Time taken: {time()-start}"
```

```python
start = time()
df["encrypted"] = df["number"].apply_parallel(encrypt_entries, num_processes=8)
f"Time taken: {time()-start}"
```
