```python
risky_operator_df = pd.read_excel("MASTER - updated 3101 2026 - risky_teleoperator_network 2.xlsx")
risky_operator_df.columns
```

```python
risky_operator_df.network_operator_name = risky_operator_df.network_operator_name.apply(lambda x: x.lower().replace(" ", "") if isinstance(x, str) else x)
risky_operators_list = risky_operator_df.network_operator_name.unique().tolist()
risky_operators_list
```

```python
network_test_df["current_network_details.name"] = network_test_df["current_network_details.name"].apply(lambda x: x.lower().replace(" ", "") if isinstance(x, str) else x)

all_available_networks = network_test_df["current_network_details.name"].unique().tolist()
all_available_networks
```

```python
remaining_networks = risky_operators_list
remaining_available_networks = all_available_networks
count = 0
for network in all_available_networks:
    global_error = True
    for risky_network in risky_operators_list:
        if isinstance(network, str) and network in risky_network:
            count += 1
            global_error = False
            remaining_networks.remove(risky_network)
            remaining_available_networks.remove(risky_network)
            break
            
    # if global_error:
    #     print(network)
remaining_networks, remaining_available_networks
```
