```python
sqs = boto3.client(
    "sqs",
    region_name="placeholder",
    aws_access_key_id="placeholder",
    aws_secret_access_key="placeholder",
)
QUEUE_URL = "https://sqs.eu-north-1.amazonaws.com/813519525710/voluum-live-data-queue"
```

```python
all_messages = []
```

```python
def poll():
    count = 0
    empty_receives = 0
    max_empty = 3  # stop after 3 consecutive empty responses

    while True:
        resp = sqs.receive_message(
            QueueUrl=QUEUE_URL,
            MaxNumberOfMessages=10,
            WaitTimeSeconds=20,
            VisibilityTimeout=60,
        )

        messages = resp.get("Messages", [])
        if not messages:
            empty_receives += 1
            if empty_receives >= max_empty:
                print("Queue empty, stopping.")
                break
            continue

        empty_receives = 0  # reset on successful receive

        for msg in messages:
            try:
                count += 1
                if count % 5000 == 0:
                    print(count)
                body = json.loads(msg["Body"])
                raw_events = json.loads(body["Message"])

                for event in raw_events:
                    all_messages.append(event)

            except Exception as e:
                print("Error:", e)
```

```python
poll()
```
